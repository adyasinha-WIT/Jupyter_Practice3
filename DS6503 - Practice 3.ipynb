**Q1 : Plot a line graph showing the trend of passenger counts over years using the `flights` dataset from Seaborn.**

**Answer**:

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Check the first few rows of the dataset to understand its structure
print(flights.head())

# Create a pivot table to aggregate data by year
flights_pivot = flights.pivot_table(index='year', values='passengers', aggfunc='sum')

# Setting the plot style
flights = sns.set_style("dark")

# Setting the plot context
flights = sns.set_context("paper")

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(flights_pivot.index, flights_pivot['passengers'], marker='o', linestyle='-', color='b')
plt.title('Trend of Passenger Counts Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

**Q10: Create a grouped bar plot to compare average miles per gallon (mpg) of cars by origin and number of cylinders using the `mpg` dataset from Seaborn.**

**Answer**:

```python
# Load the mpg  dataset
mpg  = sns.load_dataset('mpg')

# Check the first few rows of the dataset to understand its structure
print(mpg.head())

# Calculating the average mpg by origin and number of cylinders
avg_mpg = mpg.groupby(['origin', 'cylinders'])['mpg'].mean().reset_index()

# Plotting the grouped bar chart
sns.barplot(x='origin', y='mpg', hue='cylinders', data=avg_mpg, palette='rocket')
plt.title('Average MPG by Origin and Number of Cylinders')
plt.xlabel('Origin')
plt.ylabel('Average MPG')
plt.legend(title='Cylinders')
```

## 3. Interactive Data Visualization

**Q1: Load the `iris` dataset from `seaborn` and create an interactive scatter plot using `bokeh` to visualize the relationship between two numerical variables.**

**Answer**:

```python
import pandas as pd
import seaborn as sns
from bokeh.plotting import figure, show, output_notebook, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.io import output_file
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10

# Load dataset
df = sns.load_dataset('iris')

# Prepare the data
source = ColumnDataSource(df)

# Create the plot
p = figure(title="Interactive Scatter Plot of Sepal Length vs Sepal Width",
           x_axis_label='Sepal Length', y_axis_label='Sepal Width',
           tools="pan,wheel_zoom,box_zoom,reset,save")

# Define a color map
species = df['species'].unique().tolist()
color_map = factor_cmap(field_name='species', palette=Category10[3], factors=species)

# Add scatter plot
p.scatter('sepal_length', 'sepal_width', size=10, source=source, legend_field='species',
          fill_alpha=0.6, line_color=None, color=color_map)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Species", "@species"), ("Sepal Length", "@sepal_length"), ("Sepal Width", "@sepal_width")]
p.add_tools(hover)

# Customize legend
p.legend.title = 'Species'
p.legend.location = 'top_left'
p.legend.click_policy = 'hide'

# Show the plot
output_notebook()
output_file("interactive_scatter.html")
show(p)      
```

**Q2: Load the `flights` dataset from seaborn and create an interactive line plot using `bokeh` to visualize the trend over time.**

**Answer**:

```python
import pandas as pd
import seaborn as sns
from bokeh.plotting import figure, show, output_file, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.io import output_notebook

# Load dataset
df = sns.load_dataset('flights')

# Prepare the data
df['time'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-01')
source = ColumnDataSource(df)

# Create the plot
p = figure(title="Interactive Line Plot of Airline Passengers Over Time",
           x_axis_label='Time', y_axis_label='Number of Passengers',
           x_axis_type='datetime', tools="pan,wheel_zoom,box_zoom,reset,save")

# Add line plot
p.line('time', 'passengers', source=source, line_width=2)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Time", "@time{%F}"), ("Passengers", "@passengers")]
hover.formatters = {'@time': 'datetime'}
p.add_tools(hover)

# Show the plot
output_notebook()
output_file("interactive_line.html")
show(p)
```
